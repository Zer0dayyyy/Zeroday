```python
import pandas as pd
data = pd.read_csv('Hotels.csv')
df = pd.DataFrame(data)
RANGE = df[(df['price'] >= 200) & (df['price'] <= 300)]['name']
print(RANGE)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('Hotels.csv')
df = pd.DataFrame(data)

x = df['days']
y = df['total']

plt.bar(x, y)
plt.xlabel('Days')
plt.ylabel('Total')
plt.title('Days vs Total')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('Hotels.csv')
df = pd.DataFrame(data)
x = df['price']
y = df['place']
plt.bar(x, y)
plt.xlabel('price')
plt.ylabel('place')
plt.title('price vs place')

plt.show()
```
